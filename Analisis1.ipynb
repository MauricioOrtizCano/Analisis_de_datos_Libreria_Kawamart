```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pyp
```

```python
compras = pd.read_csv("Compras.csv")
```

```python
compras['Fecha del pedido'] = pd.to_datetime(compras['Fecha del pedido'], origin='1899-12-30', unit='D')
```

```python
compras['Fecha de envío'] = pd.to_datetime(compras['Fecha de envío'], origin='1899-12-30', unit='D')
```

```python
compras.head(5)
```

```python
compras['Año'] = compras['Fecha del pedido'].dt.year

compras.head(10)
```

```python
compras['Año'].value_counts(ascending=True)
```

```python
devoluciones = pd.read_csv('Devoluciones.csv')
```

```python
devoluciones.head(5)
```

```python
#Se cruzan las 2 bases de datos

compras = pd.merge(compras, devoluciones, how='left', left_on='Id. del pedido', right_on='Id. del pedido')
```

```python
compras.head(5)
```

```python
compras['Devuelto'] = compras['Devuelto'].fillna('No')

compras.head(15)
```

```python
personas = pd.read_csv('Personas.csv')
```

```python
personas.head()
```

```python
compras = pd.merge(compras, personas, how='left', left_on='Región', right_on='Región')

compras.head()
```

## 1. ¿Cómo van las ventas de la compañía en los últimos años?

```python
# Para aquellas compras que fueron devueltas es necesario dejarlas con valor $0, ya que se tuvo que haber devuelto el dinero de dicha compra

compras['Ganancia Ajustada'] = np.where(compras['Devuelto'] == 'Sí', 0, compras['Ganancia ($ COP)'])

compras.head(15)
```

```python
pd.options.display.float_format = "{:.3f}".format
```

```python
# Una vez calculada la ganancia ajustada podemos determinar el valor real de la ganancia que se obtuvo en cada año

compras.groupby('Año')['Ganancia Ajustada'].sum()
```

```python
compras.groupby('Año')['Ganancia Ajustada'].sum().plot(kind='bar', figsize=(9,5))
```

### Podemos ver que los últimos 2 años (2015 y 2016) la compañia obtuvo las ganancias más altas, debido a la pendiente positiva de la grafica podemos decir que van en ascenso las ventas y se puede pronosticar que para el año 2017 es probable que aumenten con respecto a las del año 2016
